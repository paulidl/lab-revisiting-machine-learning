### Check for null values in the numerical columns.

```python
# Function that divides data into numerical and categorical.
def numerical_and_categorical(x):
    categorical = x.select_dtypes(include = np.object)
    numerical = x.select_dtypes(include = np.number)
    return numerical, categorical
```

```python
numerical, categorical = numerical_and_categorical(data)
```
